# First lets get the algorithm working for a single reaction, with sampling of kinetics at each time step to create a plot over time

```python
from sympy import *
import numpy as np
import matplotlib.pyplot as plt
```

```python
# lets start with A + B -> C (one way reaction)

stoich_vector = Matrix([-1,-1,1]) # describes what happens to the species pops if this reactin fires

propensity_function = 0.1 # since we have only one reaction, we just define a firing rate outright => this is 20% for an A and a B to combine per unit time

# initial conditions
NA, NB, NC = 100, 80, 0                              # initial count of each species
t, T = 0.0, 500.0                               # initial and final time
times, A_counts, B_counts, C_counts = [t], [NA], [NB], [NC]

# algorithm:

while t < T:

    a0 = propensity_function*NA*NB # calculate the prob/time for reaction to occur at the current system state (defined by NA,NB)

    if a0 == 0:
        break

    r1, r2 = np.random.rand(2) # draw two random numbers - how random are they, use qiskit for quanutm randomness?

    tau = 1/a0 * np.log(1/r1)

    t += tau

    NA += stoich_vector[0,0]; NB += stoich_vector[1,0]; NC += stoich_vector[2,0] # population change depending on stoichiometric vector
    times.append(t); A_counts.append(NA); B_counts.append(NB); C_counts.append(NC)

plt.step(times, A_counts, where="post", label="A")
plt.step(times, B_counts, where="post", label="B")
plt.step(times, C_counts, where="post", label="C")
plt.xlabel("time"); plt.ylabel("Molecule count")
plt.grid(True)
plt.title("Gillespie simulation of single reaction")
plt.legend(); plt.show()

print(r'Final A: {}'.format(A_counts[-1]))
print(r'Final B: {}'.format(B_counts[-1]))
print(r'Final C: {}'.format(C_counts[-1]))
```
